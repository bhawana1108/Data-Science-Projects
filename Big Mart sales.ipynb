```python
#mapping the binary features
train_df['Item_Fat_Content'] = train_df['Item_Fat_Content'].map({'Low Fat':1, 'Regular':0})
```

```python
#create the dummy for rest variable
ItemType = pd.get_dummies(train_df['Item_Type'], prefix='ItemType', drop_first=True)
train_df=pd.concat([train_df, ItemType], axis=1)

OutletId = pd.get_dummies(train_df['Outlet_Identifier'], prefix='OutletId', drop_first=True)
train_df = pd.concat([train_df, OutletId],axis=1)

OutletEstY = pd.get_dummies(train_df['Outlet_Establishment_Year'], prefix='OutletEstY', drop_first=True)
train_df = pd.concat([train_df, OutletEstY], axis=1)

OutletSize = pd.get_dummies(train_df['Outlet_Size'], prefix='OutletSize', drop_first=True)
train_df = pd.concat([train_df, OutletSize], axis=1)

OutletLocT = pd.get_dummies(train_df['Outlet_Location_Type'], prefix='OutletLocT', drop_first=True)
train_df = pd.concat([train_df, OutletLocT], axis=1)

OutletType = pd.get_dummies(train_df['Outlet_Type'], prefix='OutletType' , drop_first=True)
train_df = pd.concat([train_df, OutletType], axis=1)
```

```python
train_df.drop(['Item_Type', 'Outlet_Identifier', 'Outlet_Establishment_Year','Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type'], axis=1, inplace=True)
```

```python
train_df.head()
```

```python
#placing the targets and inputs

inputs= train_df.drop(['Item_Identifier', 'Item_Outlet_Sales'], axis=1)

targets = train_df['Item_Outlet_Sales']
```

```python
# checking for the most important feature by extratree regressor

from sklearn.ensemble import ExtraTreesRegressor
ETR = ExtraTreesRegressor()
ETR.fit(inputs,targets)
```

```python
feature_importance = pd.Series(ETR.feature_importances_, index=inputs.columns)
feature_importance.nlargest().plot(kind='barh')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor()
RFR.fit(inputs, targets)
```

```python
feature_impor_ = pd.Series(RFR.feature_importances_ , index = inputs.columns)
feature_impor_.nlargest().plot(kind='barh')
plt.show()
```

#### Here both the models shows 'Item_MRP', 'Item_Visibility', 'OutletEstY_1998', OutletId_OUT019', 'Itm_Weight' as the best feature 

```python
#splitting the data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
target_pred = reg.predict(x_test)
```

```python
target_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

MSE = mean_squared_error(y_test,target_pred)
MAE = mean_absolute_error(y_test, target_pred)
r2 = r2_score(y_test, target_pred)
RMSE = np.sqrt(MSE)

print('Root Mean Squared Error:', RMSE)
print('Mean Absolute Error:', MAE)
print('R Squared value:', r2)
```

```python
#checking the metrics with random forest 
reg_rfr = RandomForestRegressor()
reg_rfr.fit(x_train, y_train)
```

```python
rfr_pred = reg_rfr.predict(x_test)
```

```python
rfr_pred
```

```python
#calculating its metrics

MSE = mean_squared_error(y_test, rfr_pred)
MAE = mean_absolute_error(y_test, rfr_pred)
r2 = r2_score(y_test, rfr_pred)
RMSE = np.sqrt(MSE)

print('Root Mean square error:', RMSE)
print('Mean Absolute Error:', MAE)
print('R squared Value:',r2)
```
